```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 1, 50)
y = 2 * np.sin(2 * np.pi * x)
plt.figure(figsize=(8, 4))
plt.stem(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title(' y = 2sin(2πx)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
rand_values = np.random.rand(50)
noisy_values = rand_values * y

plt.figure(figsize=(8, 4))
plt.stem(x, noisy_values)
plt.xlabel('x')
plt.ylabel('y with noisy_values')
plt.title(' y = 2sin(2πx) with noise')
plt.legend()
plt.grid(True)

plt.show()
```

```python
h_values = np.convolve(x , noisy_values)
h_values
```
