```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
f = 5
phi = 0
t_cont = np.linspace(0, 1, 1000)
y_cont = A * np.sin(2 * np.pi * f * t_cont + phi)

plt.figure(figsize=(10, 4))
plt.plot(t_cont, y_cont, label='Continuous Sinusoidal Signal', color='blue')
plt.title('Step 1: Continuous Sinusoidal Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()

f_max = f
sampling_rate = 2 * f_max
num_samples = int(sampling_rate)
t_sampled = np.linspace(0, 1, num_samples, endpoint=False)
y_sampled = A * np.sin(2 * np.pi * f * t_sampled + phi)

plt.figure(figsize=(10, 4))
plt.plot(t_cont, y_cont, label='Continuous Signal', color='blue', alpha=0.6)
plt.scatter(t_sampled, y_sampled, color='red', label='Sampled Points', zorder=5)
plt.title('Step 2: Sampling the Signal at Nyquist Rate')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()

def quantize(signal, levels):
    min_val = np.min(signal)
    max_val = np.max(signal)
    q_levels = np.linspace(min_val, max_val, levels)
    quantized_signal = np.zeros_like(signal)
    for i, s in enumerate(signal):
        idx = np.argmin(np.abs(q_levels - s))
        quantized_signal[i] = q_levels[idx]
    return quantized_signal

levels = 8
y_quantized = quantize(y_sampled, levels)

plt.figure(figsize=(10, 4))
plt.scatter(t_sampled, y_sampled, color='red', label='Sampled Signal', zorder=5)
plt.scatter(t_sampled, y_quantized, color='green', marker='x', label='Quantized Signal', zorder=5)
plt.title('Step 3: Quantized Signal (8 Levels)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
