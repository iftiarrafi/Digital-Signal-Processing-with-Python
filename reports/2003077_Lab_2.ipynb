```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-5, 6, 1)

unit_sample = np.where(n == 0, 1, 0)
unit_step = np.where(n >= 0, 1, 0)
unit_ramp = np.where(n >= 0, n, 0)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.stem(n, unit_sample, basefmt=" ")
plt.title('Unit Sample Signal (δ[n])')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(n, unit_step, basefmt=" ")
plt.title('Unit Step Signal (u[n])')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(n, unit_ramp, basefmt=" ")
plt.title('Unit Ramp Signal (r[n])')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# i. Addition and Multiplication
signal_addition = unit_sample + unit_step
signal_multiplication = unit_sample * unit_ramp

# ii. Shifting
shifted_sample = np.roll(unit_sample, 2)  # Shift by 2
shifted_step = np.roll(unit_step, -2)  # Shift by -2
shifted_ramp = np.roll(unit_ramp, -1)  # Shift by -1

# iii. Scaling
scaled_sample = 2 * unit_sample
scaled_step = 3 * unit_step
scaled_ramp = 0.5 * unit_ramp

# iv. Time Reversal
reversed_sample = unit_sample[::-1]
reversed_step = unit_step[::-1]
reversed_ramp = unit_ramp[::-1]

# v. Combination of shifting, scaling, and time reversal
combined_signal = 2 * np.roll(unit_sample, 1)[::-1]

# Plotting the signals
plt.figure(figsize=(12, 12))

# i. Addition and Multiplication
plt.subplot(5, 1, 1)
plt.stem(n, signal_addition, basefmt=" ")
plt.title('Addition: δ[n] + u[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(5, 1, 2)
plt.stem(n, signal_multiplication, basefmt=" ")
plt.title('Multiplication: δ[n] * r[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

# ii. Shifting
plt.subplot(5, 1, 3)
plt.stem(n, shifted_sample, basefmt=" ")
plt.title('Shifted δ[n] by 2')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(5, 1, 4)
plt.stem(n, shifted_step, basefmt=" ")
plt.title('Shifted u[n] by -2')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(5, 1, 5)
plt.stem(n, shifted_ramp, basefmt=" ")
plt.title('Shifted r[n] by -1')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

# iii. Scaling
plt.subplot(5, 1, 1)
plt.stem(n, scaled_sample, basefmt=" ")
plt.title('Scaled δ[n] by 2')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(5, 1, 3)
plt.stem(n, scaled_step, basefmt=" ")
plt.title('Scaled u[n] by 3')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

# iv. Time Reversal
plt.subplot(5, 1, 3)
plt.stem(n, reversed_sample, basefmt=" ")
plt.title('Reversed δ[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
```
