```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 1000
T = 1
f0 = 5
harmonics = [2, 3, 4]
amplitudes = [1.0, 0.5, 0.3, 0.2]

t = np.linspace(0, T, int(Fs*T), endpoint=False)

signal = amplitudes[0] * np.sin(2 * np.pi * f0 * t)
plt.plot(t, signal, label=f'{amplitudes[0]}*sin(2π*{f0}t)')

for i, h in enumerate(harmonics):
    y = amplitudes[i+1] * np.sin(2 * np.pi * f0 * h * t)
    signal += y
    plt.plot(t, y, label=f'{amplitudes[i+1]}*sin(2π*{h*f0}t)')

plt.figure(figsize=(12,12))
plt.plot(t, signal, 'k', linewidth=2, label='Sum')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Periodic Signal with Harmonics')
plt.legend()
plt.grid()
plt.show()
```
