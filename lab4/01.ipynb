```python
signal1 = np.array([1,2,3])
signal2 = np.array([4,5,6])

convolution_full = np.convolve(signal1 , signal2 , mode='full')
convolution_valid = np.convolve(signal1 , signal2 , mode='valid')
convolution_same = np.convolve(signal1 , signal2 , mode='same')

print(f"Signal 1 : {signal1}\n" )
print(f"Signal 2 : {signal2}\n" )

print(f" full mode convolution : {convolution_full}\n" )
print(f" valid mode convolution : {convolution_valid}\n" )
print(f" same mode convolution : {convolution_same}" )
```
