```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0, 2 * np.pi, 100)  
signal = np.sin(2 * t) 

autocorrelation = np.correlate(signal, signal, mode='full')


plt.figure(figsize=(12, 5))


plt.plot(t, signal, label="Sinusoidal Signal", color='blue')
plt.title("Original Sinusoidal Signal")
plt.show()
```
