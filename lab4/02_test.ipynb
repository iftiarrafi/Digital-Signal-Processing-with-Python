```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0, 2 * np.pi, 100) 
signal = np.sin(2 * t)


autocorr = np.correlate(signal, signal, mode='full')
lags = np.arange(-len(signal) + 1, len(signal)) 


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, signal, label="Sinusoidal Signal", color='blue')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Original Sinusoidal Signal")
plt.legend()
plt.grid()


plt.subplot(1, 2, 2)
plt.plot(lags, autocorr, label="Autocorrelation", color='red')
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.title("Autocorrelation of Sinusoidal Signal")
plt.legend()
plt.grid()


plt.tight_layout()
plt.show()
```
